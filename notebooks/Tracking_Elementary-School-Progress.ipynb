```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/states_all.csv')

pivot = df.pivot_table(
    index="YEAR",
    columns="STATE",
    values="ENROLL",
    aggfunc="sum"
)

pivot.plot(figsize=(10,6), marker="o")
plt.title("Enrollment Over Time by State")
plt.xlabel("Year")
plt.ylabel("Enrollment")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/states_all.csv')

subset = df[df["YEAR"] == 2022]   # <-- define subset

revenue_cols = ["FEDERAL_REVENUE", "STATE_REVENUE", "LOCAL_REVENUE"]
revenue_data = subset[revenue_cols].sum()

sns.barplot(x=revenue_data.index, y=revenue_data.values, palette="Set2")
plt.title("Revenue Sources (2022)")
plt.ylabel("Dollars")
plt.xticks(rotation=20)
plt.show()
```
